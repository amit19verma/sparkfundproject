# Getting data from companies.txt

```python
import numpy as np
import pandas as pd
```

```python
companies = pd.read_csv("companies.txt", sep="\t", encoding = "ISO-8859-1",engine='python')
companies['permalink'] = companies['permalink'].apply(lambda per:per.lower())
companies.head(20)
```

#### Get Unique Companies present in companies.txt

```python
unique_values =  companies.permalink.nunique(dropna = True)
unique_values
```

## Getting Data from rounds2.csv

```python
round2Company = pd.read_csv("rounds2.csv",encoding = "ISO-8859-1")
round2Company.rename(columns = {'company_permalink':'permalink'}, inplace = True)
round2Company['permalink']=round2Company['permalink'].apply(lambda link : link.lower())
round2Company.head(50)
```

### Get Unique Companies present in round2Company

```python
round2Company.permalink.nunique()
```

```python
companies.head(10)
```

#### Merging  round2Company and companies data 

```python
master_frame = pd.merge(round2Company, companies, how='inner', on='permalink')
```

```python
master_frame.info()
```

```python
master_frame.head(50)
```

## Data Prepration and Cleaning 

```python
#column missing value
master_frame.isnull().sum()
```

```python
#row missing value
master_frame.isnull().sum(axis=1)
```

```python
print(len(master_frame.index))
# summing up the missing values (column-wise)
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
master_frame.isnull().sum()
```

#### Treating Missing value in Columns

```python
# removing the three columns
master_frame = master_frame.drop('funding_round_code',axis=1)
master_frame = master_frame.drop('founded_at',axis=1)
master_frame = master_frame.drop('state_code',axis=1)
master_frame = master_frame.drop('region',axis=1)


round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
master_frame.head(20)
```

### Treating Missing Value in Rows

```python
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]
master_frame = master_frame[master_frame['country_code'].notnull()]
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
# retaining the rows having <= 5 NaNs
master_frame = master_frame[:].replace(np.NaN, 'NA')
master_frame = master_frame[(master_frame.isnull().sum(axis=1)) <= 4]
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
master_frame.head(35)
```

```python
master_frame_by_funding = master_frame.groupby('funding_round_type')
master_frame_by_funding.sum()
```

```python
#Average Funding for angel type
round(master_frame_by_funding['raised_amount_usd'].mean()['angel'])
```

```python
#Average Funding for venture type
round(master_frame_by_funding['raised_amount_usd'].mean()['venture'])
```

```python
#Average Funding for seed type
round(master_frame_by_funding['raised_amount_usd'].mean()['seed'])
```

```python
#Average Funding for private_equity type
round(master_frame_by_funding['raised_amount_usd'].mean()['private_equity'])
```

```python
#funding data based on range 5-15M
#Filtering rows which does not fell between 5to 15 million
master_frame_inrange= master_frame[(master_frame.raised_amount_usd >= 500000) & (master_frame.raised_amount_usd <= 1500000)]
round(master_frame_inrange.groupby('funding_round_type').mean())
```

## Sector Analysis

#### Create Primary sector from Category list and get mapping info

```python
master_frame['PrimarySector'] = master_frame['category_list'].apply(lambda cat : str(cat).split('|')[0] if '|' in str(cat) else cat)
master_frame.head()
```

```python
master_frame.info()
```

```python
mapping = pd.read_csv("mapping.csv", encoding = "ISO-8859-1")
mapping.info()
```

```python
mapping.set_index('category_list').head()
```

##### Mapping category_list with main sector

```python
sector_mapping = pd.melt(mapping, id_vars =['category_list'], value_vars =['Manufacturing','Automotive & Sports',
                                                              'Cleantech / Semiconductors','Entertainment',
                                                             'Health','News, Search and Messaging','Others',
                                                             'Social, Finance, Analytics, Advertising']) 
```

```python
sector_mapping = sector_mapping.loc[~(sector_mapping["value"]==0) ]
sector_mapping=sector_mapping.drop(columns="value")
```

```python
sector_mapping.rename(columns = {'variable':'main_sector'}, inplace = True)
sector_mapping.rename(columns = {'category_list':'PrimarySector'}, inplace = True)
```

```python
master_frame = pd.merge(master_frame, sector_mapping, how='inner', on='PrimarySector')
```

## Plotting

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
avg_funding = master_frame_by_funding.mean()
fraction = master_frame_by_funding['raised_amount_usd']
avg_funding
```

```python
total_investment_by_type = master_frame_by_funding.sum()
total_investment_by_type
```

```python
total_invest = master_frame['raised_amount_usd'].sum()
```

```python
total_investment_by_type['Fraction'] = total_investment_by_type['raised_amount_usd'].apply(lambda amount : amount/total_invest)

```

```python
total_investment_by_type=total_investment_by_type.reset_index()
total_investment_by_type = total_investment_by_type[total_investment_by_type['funding_round_type'].isin(['angel','seed','venture','private_equity'])]
total_investment_by_type
```

```python
# increase figure size 
plt.figure(figsize=(20, 6))

# subplot 1: statistic=mean
plt.subplot(1, 2, 1)
sns.barplot(x='raised_amount_usd', y='funding_round_type', data=total_investment_by_type)
plt.title("Average ")

#subplot 2: statistic=fraction
plt.subplot(1, 2, 2)
sns.barplot(x='Fraction', y='funding_round_type', data=total_investment_by_type)
plt.title("Fraction ")

plt.show()
```
